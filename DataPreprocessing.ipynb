```python
num_cols = df.select_dtypes(include=['number']).columns  # Select only numerical columns

# Create subplots based on the number of numerical columns
fig, axs = plt.subplots(len(num_cols), 1, dpi=95, figsize=(7, 17))

# Loop through the numerical columns and create box plots
for i, col in enumerate(num_cols):
    axs[i].boxplot(df[col], vert=False)
    axs[i].set_ylabel(col)

plt.show()
```

```python
cols = ['gas', 'std', 'two'] # drop
df = df.drop(cols, axis=1)
```

```python
df.info()
```

```python
df = df.dropna() #drop all rows in the data that have missing values (NaNs)
df.info()
```

```python
dummies = [] #Creating Dummy Variables
cols = ['four', 'rwd', 'front']
for col in cols:
   dummies.append(pd.get_dummies(df[col]))
```

```python
auto_dummies = pd.concat(dummies, axis=1)
```

```python
df = pd.concat((df,auto_dummies), axis=1)
```
